# Visulization

```python
def diabetes_category(Diabetes_012):
    diabetes_dict = {
        0: 'No Diabetes',
        1: 'Diabetes',
        2: 'Prediabetes',
       
    }
    return diabetes_dict[Diabetes_012//1]



def HighBP_category(HighBP):
    HighBP_dict = {
        0: 'No High',
        1: 'High',      
    }
    return HighBP_dict[HighBP//1]

    
def HighChol_category(HighChol):
    HighChol_dict = {
        0: 'No High',
        1: 'High',      
    }
    return HighChol_dict[HighChol//1]


def BMI_category(BMI):
    if BMI < 18.5:
        return 'under-weight'
    elif BMI >= 18.5 and BMI <= 24.9:
        return 'normal-weight'
    elif BMI >= 24 and BMI <= 29.9:
        return 'over-weight'
    elif BMI > 30.0:
        return "obese"

data['cDiabetes'] = data['Diabetes_012'].apply(lambda x: "none")
data['cHighChol'] = data['HighChol'].apply(lambda x: "none")
data['cHighBP'] = data['HighBP'].apply(lambda x: "none")
data['cBMI'] = data['BMI'].apply(lambda x: "none")
for idx, row in data.iterrows():
    data.at[idx, 'cDiabetes'] = diabetes_category(row['Diabetes_012'])
    data.at[idx, 'cHighChol'] = HighChol_category(row['HighChol'])
    data.at[idx, 'cHighBP'] = HighBP_category(row['HighBP']) 
    data.at[idx, 'cBMI'] = BMI_category(row['BMI'])

by_Diabetes = data.groupby(by='cDiabetes').size()
print (by_Diabetes)
print ('-'*80)

by_HighChol = data.groupby(by='cHighChol').size()
print (by_HighChol)
print ('-'*80)

by_HighBP = data.groupby(by='cHighBP').size()
print (by_HighBP)
print ('-'*80)

by_BMI = data.groupby(by='cBMI').size()
print (by_BMI)
print ('-'*80)
```

To understand correlation between the input or the independent variables, I classified the Diabetes,HighBP, HighChol, BMI into specific categories.  Taking the BMI ranges from Medline Plus, I categorized the input samples into 'below-weight', 'normal-weight', 'over-weight' and 'obese'

```python
variables = ['cDiabetes','cHighChol','cHighBP','cBMI','Smoker','Sex','Age']

# data distribution analysys
print('Data distribution analysys')
for v in variables:
    data = data.sort_values(by=[v])
    data[v].value_counts().plot(kind = 'bar')
    plt.title(v)
    plt.show()
```

From the above data we can observe below

# Creating independent and dependent variable

```python
X =data.drop(['Diabetes_012','cDiabetes','cHighChol','cHighBP','cBMI'], axis=1)
y = data['Diabetes_012'].values
```

# Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3,random_state = 1)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_train = SS.fit_transform(X_train)
X_test = SS.transform(X_test)
```

# K-Nearest Neighbor(KNN) Model

```python
#Fitting K-NN classifier to the training set  
from sklearn.neighbors import KNeighborsClassifier  
KNclassifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
KNclassifier.fit(X_train, y_train)
```

```python
#Predicting the test set result  
ypred= KNclassifier.predict(X_test) 
```

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
CM= confusion_matrix(y_test, ypred)
AC = accuracy_score(y_test,ypred)
plot_confusion_matrix(KNclassifier,X_test, y_test)
```

```python
print(CM)
```

```python
print(classification_report(y_test,ypred))
```
